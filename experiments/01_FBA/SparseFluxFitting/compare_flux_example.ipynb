```python
import pandas as pd

df_0 = pd.read_csv("prediction_lam0.csv", index_col=0)
df_1 = pd.read_csv("prediction_lam1.csv", index_col=1)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2, figsize=(8, 4), sharey=True)
df_0.plot.scatter(x="true", y="pred", ax=ax[0])
df_1.plot.scatter(x="true", y="pred", ax=ax[1])
ax[0].set_xlabel("true flux")
ax[0].set_ylabel("predicted flux")
ax[0].set_title("$\\lambda=0$, test fraction = 0.8")
plt.tight_layout()
ax[1].set_xlabel("true flux")
ax[1].set_ylabel("predicted flux")
ax[1].set_title("$\\lambda=1$, test fraction = 0.8")
plt.savefig(f"compare_lam0_lam1_test_frac_0.8_noise_1.0.pdf")
```
